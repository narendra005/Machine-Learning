```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
type(boston)
```

```python
df_x=pd.DataFrame(boston.data,columns=boston.feature_names)
df_y=pd.DataFrame(boston.target)
```

```python
df_x
```

```python
## df_x is the Main Dataframe and df_y is dataframe with only one Column and Different value likely to be output variable

reg =linear_model.LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, random_state=4,test_size=.8)
```

```python
print("The No of Rows Columns in X_Train is which is based on test_size given earlier",X_train.shape)
print("The No of Rows Columns in X_test is which is based on test_size given earlier",X_test.shape)
print("The No of Rows Columns in y_train is which is based on test_size given earlier",y_train.shape)
print("The No of Rows Columns in y_test is which is based on test_size given earlier",y_test.shape)
```

```python
### Step to Calculate Regerssion
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
```

```python
#########Pridicting the Value 
a=reg.predict(X_test)
Y_test=np.array(y_test.values)
```
