```python
import fitz
import tabula

pdf_document = (r"C:\Users\Dell\Downloads\animesh\0011164676-296-F8U-AGT.pdf")

df = tabula.read_pdf(pdf_document,pages='all',stream = True)

doc = fitz.open(pdf_document)
print ("number of pages: %i" % doc.pageCount)
print(doc.metadata)

page1 = doc.loadPage(0)
page1text = page1.getText("blocks")
#print(page1text)
ctr=0
fln=' '
lno=1
arr1=[]

for ln in page1text :
    print(">>",ln)
   
    (x0, y0, x1, y1, bvar, linno, wrdno) = ln
    bvar = bvar.replace("\n","  ♀ ")
    print("\t\t>>",linno,",",wrdno,">>",bvar)
   
    if linno > 4:   
        arr1.append([])
        arr1[ctr].append(x0)
        arr1[ctr].append(y0)
        arr1[ctr].append(x1)
        arr1[ctr].append(y1)
        arr1[ctr].append(bvar)
        arr1[ctr].append(linno)
        arr1[ctr].append(wrdno)
        ctr += 1
print(df)
arr1
```
